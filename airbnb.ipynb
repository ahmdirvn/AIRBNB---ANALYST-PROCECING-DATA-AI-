# 3 # Pembersihan data menggunakan pandas

```python
#mencari jumlah missing value 
df.isnull().sum()
```

```python
#mengatasi missing value dengan nilai lain
df['name'].fillna('Unknown', inplace=True) 
df['host_name'].fillna('Unknown',inplace = True)
reviews_per_month_med = df['reviews_per_month'].median
df['last_review'].fillna('Unknown', inplace=True)
df['reviews_per_month'].fillna(reviews_per_month_med, inplace=True)
```

# 5 # VISUALISASI DATA

```python
import matplotlib.pyplot as plt

# Plot histogram of price
plt.hist(df['price'], bins=30, edgecolor='k')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data untuk neighbourhood_group Manhattan
manhattan_data = df[df['neighbourhood_group'] == 'Manhattan'].nlargest(10, 'price')

# Filter data untuk neighbourhood_group Brooklyn
brooklyn_data = df[df['neighbourhood_group'] == 'Brooklyn'].nlargest(10, 'price')

# Membuat line chart untuk 10 harga teratas di neighbourhood_group Manhattan dengan warna merah (red)
plt.plot(manhattan_data['price'], marker='o', color='red', label='Manhattan')

# Membuat line chart untuk 10 harga teratas di neighbourhood_group Brooklyn dengan warna biru (blue)
plt.plot(brooklyn_data['price'], marker='o', color='blue', label='Brooklyn')

# Menentukan label pada sumbu x
plt.xticks(range(len(manhattan_data)), manhattan_data['name'], rotation=45)

# Menentukan label pada sumbu y
plt.ylabel('Price')

# Memberikan judul pada line chart
plt.title('Perbandingan 10 Harga Teratas di Neighbourhood Group Manhattan dan Brooklyn')

# Menampilkan legend untuk menunjukkan keterangan garis
plt.legend()

# Menampilkan line chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Menghitung jumlah masing-masing kategori room_type
room_type_counts = df['room_type'].value_counts()

# Mendapatkan label dan nilai jumlah room_type
labels = room_type_counts.index
sizes = room_type_counts.values

# Membuat pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Menambahkan judul pada pie chart
plt.title('Persentase Jumlah Room Type')

# Menampilkan pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Mengurutkan dataframe berdasarkan kolom 'number_of_reviews' secara descending
sorted_data = df.sort_values('number_of_reviews', ascending=False)

# Mengambil 10 nilai tertinggi dalam kolom 'number_of_reviews'
top_10_reviews = sorted_data.head(10)

# Menentukan posisi bar pada sumbu x
x = range(len(top_10_reviews))

# Menampilkan bar chart
plt.bar(x, top_10_reviews['number_of_reviews'])

# Menentukan label pada sumbu x dengan ukuran font yang lebih kecil
plt.xticks(x, top_10_reviews['name'], rotation=45, fontsize=8)

# Menentukan label pada sumbu y dengan ukuran font yang lebih kecil
plt.ylabel('Number of Reviews', fontsize=8)

# Memberikan judul pada bar chart
plt.title('Top 10 Listings with Highest Number of Reviews')

# Menampilkan bar chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Mengambil subset dataset dengan variabel 'room_type' dan 'number_of_reviews'
subset_data = df[['room_type', 'number_of_reviews']]

# Mengambil data dengan jumlah review 10 atau lebih
filtered_data = subset_data[subset_data['number_of_reviews'] >= 10]

# Mengurutkan data berdasarkan jumlah review secara menurun
sorted_data = filtered_data.sort_values(by='number_of_reviews', ascending=False)

# Mengambil 10 data dengan jumlah review terbanyak
top_10_data = sorted_data.head(10)

# Membuat visualisasi bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_10_data['room_type'], top_10_data['number_of_reviews'])

# Memberikan label pada sumbu-x dan sumbu-y
plt.xlabel('Number of Reviews')
plt.ylabel('Room Type')

# Memberikan judul pada chart
plt.title('Top 10 Room Types by Number of Reviews')

# Menampilkan chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Mengambil subset dataset dengan variabel 'price'
subset_data = df[['price']]

# Mengurutkan data berdasarkan harga secara menurun
sorted_data = subset_data.sort_values(by='price', ascending=False)

# Mengambil 10 data harga termahal
top_10_data = sorted_data.head(10)

# Membuat line chart
plt.figure(figsize=(10, 6))
plt.plot(top_10_data.index, top_10_data['price'], marker='o')

# Memberikan label pada sumbu-x dan sumbu-y
plt.xlabel('Index')
plt.ylabel('Price')

# Memberikan judul pada chart
plt.title('Top 10 Highest Prices')

# Menampilkan chart
plt.show()
```
